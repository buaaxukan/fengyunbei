```python
import numpy as np
import time

def change_timestamp(dt):

    timeArray = time.strptime(dt, "%Y-%m-%d %H:%M:%S")
    timestamp = int(time.mktime(timeArray))
    return timestamp
def change_timestamp2(dt):
    timeArray = time.strptime(dt, "%Y-%m-%d")
    timestamp = int(time.mktime(timeArray))
    return timestamp
```
